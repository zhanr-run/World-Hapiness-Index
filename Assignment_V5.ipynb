### Remove useless variables and rename columns

We can see these dataset has different columns and even the same variable have different name, so we rename them in order to combine them.

After looking the details of the each column, there are 9 common variables in thses datasset, namely, country,  happiness rank, happiness score, GDP per capital, social support, health(life expectancy), freedom(to choose life), government corruption, generosity.

Feature：If you add all these factors up, you get the happiness score so it might be un-reliable to model them to predict Happiness Scores.

```python
# drop and rename the common variables
df15 = df15.drop(['Standard Error','Dystopia Residual'],axis=1)
df15 = df15.rename(columns ={'Happiness Rank' : 'Rank_15', 'Happiness Score' : 'Score_15','Economy (GDP per Capita)' : 'Economy_15', 
                             'Family':'Family_15','Health (Life Expectancy)': 'Health_15', 'Freedom':'Freedom_15','Generosity':'Generosity_15',
                             'Trust (Government Corruption)' : 'Corruption_15'})
df15['Year'] = 2015

df16 = df16.drop(['Dystopia Residual', 'Lower Confidence Interval', 'Upper Confidence Interval'],axis=1)
df16 = df16.rename(columns ={'Happiness Rank' : 'Rank_16', 'Happiness Score' : 'Score_16','Economy (GDP per Capita)' : 'Economy_16', 'Family':'Family_16','Health (Life Expectancy)': 'Health_16',
                             'Freedom':'Freedom_16','Generosity':'Generosity_16','Trust (Government Corruption)' : 'Corruption_16'})
df16['Year'] = 2016

df17 = df17.drop(['Whisker.high', 'Whisker.low', 'Dystopia.Residual'],axis=1)
df17 = df17.rename(columns ={'Happiness.Rank': 'Rank_17', 'Happiness.Score' : 'Score_17', 'Economy..GDP.per.Capita.' : 'Economy_17', 'Family':'Family_17','Health..Life.Expectancy.': 'Health_17', 
                             'Freedom':'Freedom_17','Generosity':'Generosity_17','Trust..Government.Corruption.': 'Corruption_17'})
df17['Year'] = 2017

# df18 and df19 have the exact 9 common columns, we only need to rename them in the original order
df18 = df18.rename(columns = {'Overall rank': "Rank_18",'Country or region':'Country', 'Score':"Score_18",'GDP per capita':'Economy_18',
                          'Social support':'Family_18','Healthy life expectancy':'Health_18','Freedom to make life choices':'Freedom_18',
                          'Generosity':'Generosity_18','Perceptions of corruption':'Corruption_18'})
df18['Year'] = 2018

df19= df19.rename(columns = {'Overall rank': "Rank_19",'Country or region':'Country', 'Score':"Score_19",'GDP per capita':'Economy_19',
                          'Social support':'Family_19','Healthy life expectancy':'Health_19','Freedom to make life choices':'Freedom_19',
                          'Generosity':'Generosity_19','Perceptions of corruption':'Corruption_19'})
df19['Year'] = 2019
```

### Adding Region Column

Region is a key factor in our case, however the df17, df18, df19 do not include the Region, so we need to add a region column for them (based on the df15,df16).

```python
# define a function to create Region column for df17,df18, df19.
def add_region_column(df):
    # firstly create a dictionary from df15 with country as key, region as value
    dict15 = dict(zip(df15['Country'],df15['Region']))
    
    # create a new region column for the dataset based on the dict15,if there is no reference country we fill in NaN
    df['Region'] = df['Country'].apply(lambda x: dict15[x] if x in dict15 else NaN )
    
    # then create a same dictionaty from df16 
    dict16 = dict(zip(df16['Country'],df16['Region']))
    
    # fill the region if the country in df also show in df16, if there is no reference country we leave as original
    df['Region'] = df['Region'].apply(lambda x: dict16[x] if x in dict16 else x )
    return df.head(3)
```

```python
for dataframe in [df17,df18,df19]:
    add_region_column(dataframe)    
```

## Choosing Colors

```python
#creat the palette for further visualization
def sns_palette(A,B,n):
    palette = []
    D = [(B[i]-A[i])/255/(n-1) for i in range(len(A))]
    start = [A[i]/255 for i in range(len(A))]
    for i in range(n):
        palette.append((start[0]+i*D[0],start[1]+i*D[1],start[2]+i*D[2]))
    return palette
def pd_palette(A,B,n):
    palette = []
    D = [(B[i]-A[i])/255/(n-1) for i in range(len(A))]
    start = [A[i]/255 for i in range(len(A))]
    for i in range(n):
        palette.append([start[0]+i*D[0],start[1]+i*D[1],start[2]+i*D[2],1.])
    return np.array(palette)
#choosing rgb range from (131,186,12),(251,231,37)
palette_ = sns_palette((131,186,12),(251,231,37),20)
sns.palplot(palette_)
plt.savefig("colorbar.png")
```
